```python
import tensorflow as tf
import numpy as np
import os
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.preprocessing import image
```

### EĞİTİLMİŞ BİR GÖRSEL TANIMLAMA MODELİ OLAN VGG16'NIN AĞIRLIKLARININ PROJEYE DAHİL EDİLMESİ

```python
model = VGG16(weights='imagenet', include_top=False)
```

### GÖRSEL YÜKLEME FONKSİYONUNUN OLUŞTURULMASI VE GÖRSEL BOYUTUNUN BELİRTİLMESİ

```python
def load_image(img_path):
    img = image.load_img(img_path, target_size=(200, 200))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    return x
```

### GÖRSELLERİN DOSYADAN PROJEYE YÜKLENMESİ

```python
image_directory = "C:/Users/berka/Masaustu/YapayZeka/ImagePredict/Images"
images = []

for filename in os.listdir(image_directory):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        img_path = os.path.join(image_directory, filename)
        img = load_image(img_path)
        images.append(img)

images = np.vstack(images)
```

```python
resim_sec = "./Images/33.png"
selected_image = load_image(resim_sec)
```

### HAZIR EĞİTİLMİŞ MODELİ KULLANARAK TAHMİNDE BULUNMA

```python
selected_image_features = model.predict(selected_image)
image_features = model.predict(images)

similarities = tf.keras.losses.cosine_similarity(selected_image_features, image_features).numpy().flatten()
```

### ÖRNEKTEKİ RESME EN ÇOK BENZEYEN 10 RESMİ ÇIKTI OLARAK YAZDIRMA (ÇOKTAN AZA)

```python
import matplotlib.pyplot as plt

sorted_indices = np.argsort(similarities)[::-1]

for i in range(10):
    similar_image_index = sorted_indices[i]
    similar_image_path = os.listdir(image_directory)[i]
    similarity_score = similarities[similar_image_index]
    print("Benzerlik Oranı:", similarity_score)
    similar_image = image.load_img(os.path.join(image_directory, similar_image_path))
    plt.imshow(similar_image)
    plt.axis('off')
    plt.show()
```
